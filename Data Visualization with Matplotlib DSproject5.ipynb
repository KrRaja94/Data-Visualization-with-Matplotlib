```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

#### BASIC GRAPH

```python
X = [0,1,2,5]
Y = [4,2,4,10]

#re-size the graph (dpi specifies pixel per inch)
plt.figure(figsize=(8,4), dpi=100)
plt.plot(X,Y, label='2x', linewidth=12 , linestyle="--" , marker='o', markersize=20, mfc="red", mec="black") #plot graph lineraly 

plt.title("My first graph" , fontdict={"fontname": "times new roman" , 'fontsize':20, 'color':"red"}) 
#to give the title to the graph along with the font design 

plt.xlabel('X axis')
plt.ylabel('Y axis')

#plot line 2
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:6], x2[:6]**2 , 'black' , label="x^2")
plt.plot(x2[5:], x2[5:]**2 , '--')
plt.xticks([0,1,2,3,4,5,], label="avd") #to give the duration over the axis 
plt.yticks([0,2,4,6,8,12,14])

#to save a graph
plt.savefig("mygig.jpg", dpi=300)

plt.show() 
```

### BAR CHAT

```python
labels = ['a', 'b', 'c']
values = [1,4,2]

bars = plt.bar (labels, values)
#for bar design 


#2nd method 
patterns = ['|', '*', '+']
for bar in bars:
    bar.set_hatch(patterns.pop())
```
